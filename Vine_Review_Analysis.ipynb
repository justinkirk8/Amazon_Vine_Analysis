```python
# Add dependencies.
import pandas as pd
import os
```

```python
# File path
data_to_load = os.path.join("vine_table.csv")
```

```python
#create initial dataframe
vine_df = pd.read_csv(data_to_load)
vine_df.head()
```

```python
#filter total_votes > 20
vine_20votes_df = vine_df.loc[vine_df['total_votes'] > 20]
vine_20votes_df.head()
```

```python
#filter [ helpful_votes / total_votes ] > 0.5.  Step 1 add calculation in row, step 2 filter row
vine_20votes_df ['temp'] = vine_20votes_df['helpful_votes']/vine_20votes_df['total_votes']
vine_helpful_df = vine_20votes_df.loc[vine_20votes_df['temp']>0.5]
vine_helpful_df.head()
```

```python
#drop temp row
vine_helpful_df = vine_helpful_df.drop(['temp'],axis=1)
vine_helpful_df.head()
```

```python
#df for all reviews part of the vine program
vine_Y_df = vine_helpful_df.loc[vine_helpful_df['vine'] == 'Y']
vine_Y_df.head()
```

```python
#df for all reviews part of the vine program
vine_N_df = vine_helpful_df.loc[vine_helpful_df['vine'] == 'N']
vine_N_df.head()
```

```python
#determining requested stats for Y group
# total count
print(f"Vine Reviews")

y_count = len(vine_Y_df)
print(f"Number of reviews: {y_count}")

# number of 5 star
y_5_star = len(vine_Y_df.loc[vine_Y_df['star_rating'] == 5])
print(f"Number of 5 star reviews: {y_5_star}")

# % of five star reviews
y_5_star_per = round(y_5_star / y_count * 100)
print(f"Percentage of 5-star reviews: {y_5_star_per}%")

```

```python
#determining requested stats for N group
# total count
print(f"Non-Vine Reviews")

n_count = len(vine_N_df)
print(f"Number of reviews: {n_count}")

# number of 5 star
n_5_star = len(vine_N_df.loc[vine_N_df['star_rating'] == 5])
print(f"Number of 5 star reviews: {n_5_star}")

# % of five star reviews
n_5_star_per = round(n_5_star / n_count * 100)
print(f"Percentage of 5-star reviews: {n_5_star_per}%")
```
